```python
#P6
import math
def binary(x):
  if(x<0):
    return 0
  else:
    return 1
def sigmoid(x):
  return 1/(1+math.exp(-x))
def elu(x):
  if(x<0):
    return 0.01*(math.exp(x)-1)
  else:
    return x;
def is_number(x):
    try:
        float(x)
        return True
    except ValueError:
        return False
x =    input ("Nhap x = ")
if(is_number(x)==False):
  print("x must be a number")
else:
  x = float(x)
  while True:
    ham = input("Nhap ham (binary | sigmoid | elu) : ")
    if(ham == "binary" or ham == "sigmoid" or ham == "elu"):
      break;
    else:
      print(f"{ham} is not supported. ")
  if(ham == "binary"):
   print(binary(x))
  elif(ham == "sigmoid"):
   print(sigmoid(x))
  elif(ham == "elu"):
   print(elu(x))
```

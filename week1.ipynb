Câu 2: Hãy so sánh đặc điểm của Python với ngôn ngữ C++

Python là một ngôn ngữ lập trình bậc cao, thông dịch (interpreted), và đa mục đích. Nó được phát triển bởi Guido van Rossum và lần đầu tiên được giới thiệu vào năm 1991. Python nổi tiếng với cú pháp đơn giản, dễ đọc và dễ hiểu, cho phép người lập trình viết mã nhanh chóng và dễ dàng.

C++ là một ngôn ngữ lập trình máy tính bậc cao, phức tạp và đa mục đích. Nó được phát triển từ ngôn ngữ C vào những năm 1980 bởi Bjarne Stroustrup tại Bell Labs, Mỹ. C++ đã trở thành một trong những ngôn ngữ lập trình phổ biến và mạnh mẽ nhất trên thế giới và thường được sử dụng trong nhiều lĩnh vực công nghiệp, phát triển phần mềm, và trí tuệ nhân tạo.

Câu 3: Python là ngôn ngữ biên dịch hay thông dịch? Vì sao?

Python là một ngôn ngữ lập trình thông dịch (interpreted language) được hiểu là mã nguồn của chương trình lập trình Python được chuyển đổi thành mã byte . Bởi vì, code Python được thông dịch và thực thi từng dòng một khi chạy chương trình mà không cần trải qua quá trình biên dịch sang ngôn ngữ máy trước khi chạy.

Các toán tử:


```python
# Toán tử số học
a = 10
b = 3
print(a + b)   # Cộng: 13
print(a - b)   # Trừ: 7
print(a * b)   # Nhân: 30
print(a / b)   # Chia: 3.333...
print(a // b)  # Chia lấy phần nguyên: 3
print(a % b)   # Chia lấy phần dư: 1
print(a ** b)  # Lũy thừa: 1000

# Toán tử so sánh
print(a == b)  # Bằng: False
print(a != b)  # Khác: True
print(a > b)   # Lớn hơn: True
print(a < b)   # Nhỏ hơn: False
print(a >= b)  # Lớn hơn hoặc bằng: True
print(a <= b)  # Nhỏ hơn hoặc bằng: False

# Toán tử logic
x = True
y = False
print(x and y)  # AND: False
print(x or y)   # OR: True
print(not x)    # NOT: False

# Toán tử gán
c = 5
c += 2          # c = c + 2: 7
c -= 3          # c = c - 3: 4
c *= 2          # c = c * 2: 8
c /= 4          # c = c / 4: 2.0

# Toán tử bit
d = 5           # 5 = 101 (binary)
e = 3           # 3 = 011 (binary)
print(d & e)    # AND: 1 (001)
print(d | e)    # OR: 7 (111)
print(d ^ e)    # XOR: 6 (110)
print(~d)       # NOT: -6 (bit đảo)

# Toán tử thành viên
fruits = ["apple", "banana", "cherry"]
print("apple" in fruits)    # True
print("grape" not in fruits) # True

# Toán tử nhận dạng
f = [1, 2, 3]
g = [1, 2, 3]
h = f
print(f is h)    # True (cùng đối tượng)
print(f is g)    # False (khác đối tượng)
print(f is not g)# True
```
